```python
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor
import torch
import soundfile as sf

# Load pretrained Wav2Vec2 model and processor
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-large-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-large-960h")
```

```python
def transcribe_audio(audio_file):
    # Read the audio file
    audio_input, sample_rate = sf.read(audio_file)
    # Process the audio for the model
    inputs = processor(audio_input, sampling_rate=sample_rate, return_tensors="pt", padding=True)
    # Get logits from the model
    logits = model(inputs.input_values).logits
    # Predict the most likely IDs
    predicted_ids = torch.argmax(logits, dim=-1)
    # Decode the IDs to text
    return processor.decode(predicted_ids[0])
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
from torch.utils.data import Dataset, DataLoader
import torch
```

```python
# Remove "Customer:" and "Salesman:" prefixes
final_df["Customer Dialogue"] = final_df["Customer Dialogue"].str.replace(r"^Customer:\s*", "", regex=True)
final_df["Salesperson Response"] = final_df["Salesperson Response"].str.replace(r"^Salesman:\s*", "", regex=True)
```

```python
final_df.head()
```

```python
class DialogueDataset(Dataset):
    def __init__(self, data, tokenizer, max_len=512):
        self.data = data
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        customer = self.data[index]["Customer Dialogue"]
        salesperson = self.data[index]["Salesperson Response"]
        
        # Tokenize inputs and outputs
        inputs = self.tokenizer(
            customer, max_length=self.max_len, padding="max_length", truncation=True, return_tensors="pt"
        )
        outputs = self.tokenizer(
            salesperson, max_length=self.max_len, padding="max_length", truncation=True, return_tensors="pt"
        )

        return {
            "input_ids": inputs["input_ids"].squeeze(),
            "attention_mask": inputs["attention_mask"].squeeze(),
            "labels": outputs["input_ids"].squeeze(),
        }
```

```python
# Initialize Tokenizer and Dataset
tokenizer = T5Tokenizer.from_pretrained("t5-small")
dataset = DialogueDataset(final_df.to_dict(orient="records"), tokenizer)
```

```python
loader = DataLoader(dataset, batch_size=4, shuffle=True)
```

```python
model = T5ForConditionalGeneration.from_pretrained("t5-small")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

```python
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
```

```python
epochs = 3
for epoch in range(epochs):
    model.train()
    for batch in loader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        optimizer.zero_grad()
        outputs = model(
            input_ids=input_ids, attention_mask=attention_mask, labels=labels
        )
        loss = outputs.loss
        loss.backward()
        optimizer.step()

    print(f"Epoch {epoch + 1} Loss: {loss.item()}")
```

```python
# Save Model
model.save_pretrained("./salesperson-response-model")
tokenizer.save_pretrained("./salesperson-response-model")
```

```python
def generate_response(model, tokenizer, text, max_len=50):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Determine the device
    model = model.to(device)  # Move model to the device

    inputs = tokenizer(text, return_tensors="pt", max_length=512, truncation=True, padding="max_length")
    input_ids = inputs["input_ids"].to(device)  # Move input tensors to the device
    attention_mask = inputs["attention_mask"].to(device)  # Move attention mask to the device

    outputs = model.generate(input_ids=input_ids, attention_mask=attention_mask, max_length=max_len)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
```

```python
customer_dialogue = "Hi, Im interested in purchasing a new smartphone"
response = generate_response(model, tokenizer, customer_dialogue)
print("Generated Response:", response)
```
